```python
import csv
import pandas as pd
import numpy as np
from fuzzywuzzy import process, fuzz

Match = pd.read_csv(r'C:\Users\sb1923\Downloads\AirQualityUCI\AirQualityUCI.csv', delimiter=';')
```

```python
for col in Match[['Date','Time','CO(GT)','PT08.S1(CO)','NMHC(GT)','C6H6(GT)','PT08.S2(NMHC)','NOx(GT)','PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH','AH']]:
    Match[col] = Match[col].astype(str).str.strip()
    print('Number of unique values in ' + str(col) +': ' + str(Match[col].nunique()))

```

```python
unique_Date = Match['Date'].unique().tolist()
sorted(unique_Date)
```

```python
score_sort = [(x,) + i
             for x in unique_Date
             for i in process.extract(x, unique_Date,     scorer=fuzz.token_sort_ratio)]
```

```python
similarity_sort = pd.DataFrame(score_sort, columns=['Date_sort','match_sort','score_sort'])
similarity_sort
```

```python
similarity_sort['sorted_Date_sort'] = np.minimum(similarity_sort['Date_sort'], similarity_sort['match_sort'])
similarity_sort
```

```python
cols_to_check = ['Date']

specified_value='nan'

# Define a rule to identify rows with NaN values
def rule(row):
    return specified_value not in row[cols_to_check].values

# Initialize an empty list to hold the rows that do not contain NaN values
new_data = []

# Iterate through each row in the DataFrame
for index, row in Match.iterrows():
    # Apply the rule defined in step 3 to the current row
    if rule(row):
        # If the rule evaluates to False, then append the current row to the "new_data" list
        new_data.append(row)

# Convert the "new_data" list to a new pandas DataFrame without NaN valued rows
new_data = pd.DataFrame(new_data, columns=Match.columns)

```
